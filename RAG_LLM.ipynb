```python
with open("python_inbuildfunctions.pdf","rb") as f:
  loader = UnstructuredFileIOLoader(file=f)
  documents = loader.load()

documents
```

```python
text_splitter = CharacterTextSplitter(chunk_size=1000,
                                      chunk_overlap=100)
texts = text_splitter.split_documents(documents)
```

```python
type(texts)
```

```python
len(texts)
```

```python
texts[2]
```

```python
embeddings = HuggingFaceEmbeddings()
```

```python
persist_directory = "vector_db"
```

```python
vectordb = Chroma.from_documents(texts, embeddings, persist_directory=persist_directory)
```

```python
retriever = vectordb.as_retriever()
```

```python
llm = ChatGroq(model = "Llama-3.3-70b-Versatile", temperature=0)
```

```python
qa_chain = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=retriever,return_source_documents=True)
```

```python
query = "what are the inbuild functions from this pdf"
```

```python
result = qa_chain.invoke({"query": query})


result["result"]
```

```python
query = 'what are the Advantage of Functions in Python give me a breif from this pdf?'
response1 = qa_chain.invoke({"query": query})
print(response1["result"])
print("*"*20)
```
